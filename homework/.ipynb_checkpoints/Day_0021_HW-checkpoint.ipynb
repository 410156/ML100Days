```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
df = sns.load_dataset('titanic')
df.info()
```

```python
sns.barplot(x = "alone", y = "fare", hue = "sex", data = df)
plt.show()
```

```python
g = sns.FacetGrid(df, col = "survived")
g.map(plt.hist,"fare")
plt.show()
```

```python
survived=df.groupby(['sex','alone']).survived.sum()
survived.plot(kind='bar')
```

```python
survived_counts = pd.crosstab([df.alone, df.sex],df.survived)
survived_counts 
```

```python
survived_counts.plot(kind='bar')
```

```python
sns.violinplot(data=survived_counts)
plt.show()
```
