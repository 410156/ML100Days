```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,720)*np.pi/180
y = np.cos(x)
plt.plot(x,y)
plt.savefig("filename.png",dpi=300,format="png")
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

T = np.arctan2(Y,X)
plt.scatter(X,Y,c=T,alpha=.5)
```
