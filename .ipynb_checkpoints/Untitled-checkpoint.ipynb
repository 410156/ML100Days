```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
store={}
for j in range(11,14):
    response = requests.get(
    "https://fubon-ebrokerdj.fbs.com.tw/z/ze/zeg/zeg_EB0%s000E_I.djhtm"%(str(j)))
    soup = BeautifulSoup(response.text, "html.parser")
    typed=soup.find("tr", class_="t01").find("td", class_="t10").select("select")[1].find_all("option")[j-11].getText()
    store[typed]={}
   # print(typed)
    test1=soup.find_all("td", class_="t3g2")
    test2=soup.find_all("td", class_="t3r2")
    
    q=0
    #for i in soup.find_all("td", class_="t3t1"):
    #    print(re.findall('\'*.....\', \'.*\'',str(i)))
    for i in range(len(soup.find_all("td", class_="t3t1"))):
        #('*','*')
        #name=soup.find_all("td", class_="t3t1")[i]
        name=re.findall('\'*.....\', \'.*\'',str(soup.find_all("td", class_="t3t1")[i]))[0]
        store[typed][name]={}
        store[typed][name]['close']=soup.find_all("td", class_="t3n1")[q].getText()
        if soup.find_all("td", class_="t3n1")[q+1].getText()=="0.00":
            q=q+2
        store[typed][name]['open']=soup.find_all("td", class_="t3n1")[q+1].getText()       
        store[typed][name]['high']=soup.find_all("td", class_="t3n1")[q+2].getText()
        store[typed][name]['low']=soup.find_all("td", class_="t3n1")[q+3].getText() 
       # while soup.find_all("td", class_="t3n1")[q+4].find(".")!="None" or soup.find_all("td", class_="t3n1")[q+4].find(".")!=[] :
       #     q=q-1
       # print(soup.find_all("td", class_="t3n1")[q+4].find("."))
        store[typed][name]['number']=soup.find_all("td", class_="t3n1")[q+4].getText() 
        store[typed][name]['limit_up']=soup.find_all("td", class_="t3n1")[q+5].getText() 
        store[typed][name]['limit_down']=soup.find_all("td", class_="t3n1")[q+6].getText() 
        q=q+7
store
```
